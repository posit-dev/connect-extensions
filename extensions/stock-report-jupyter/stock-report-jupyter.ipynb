```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

prices = pd.read_csv('gspc.csv', parse_dates=[0], index_col=0)

latest = prices.last_valid_index()
data=[
    [
        prices[latest:]['High'].values[0],
        prices[latest:]['Low'].values[0],
        prices[latest:]['Volume'].values[0].round(),
    ],
    [
        prices.asfreq('D').rolling(window=52*7, min_periods=1).max()[latest:]['High'].values[0],
        prices.asfreq('D').rolling(window=52*7, min_periods=1).min()[latest:]['Low'].values[0],
        prices.asfreq('D').rolling(window=52*7, min_periods=1).mean()[latest:]['Volume'].values[0].round(),
    ]
]

pd.DataFrame(data, columns=['High', 'Low', 'Volume'], index=['Most Recent Trading Day', '52-Week'])
```
